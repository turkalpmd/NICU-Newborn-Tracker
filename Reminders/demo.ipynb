```python
from reminders import premature_reminders, create_reminder, magic_convert_date
import random
import datetime as dt
from datetime import datetime, timedelta
import pandas as pd
import numpy as np

start_date = dt.date(2023, 3, 1)
end_date = dt.date(2023, 4, 1)

num_dates = 10#int(input("Kaç tarih oluşturmak istiyorsunuz? "))
birth_date_list = []
gestational_week_list = []
gestational_week_plus_day_list = []
for i in range(num_dates):
    random_date = start_date + (end_date - start_date) * random.random()
    random_week = random.randint(26, 41)
    random_day = random.randint(0, 6)
    #random_date += datetime.timedelta(weeks=random_week, days=random_day)
    birth_date_list.append(random_date.strftime('%Y-%m-%d'))
    gestational_week_list.append(random_week)
    gestational_week_plus_day_list.append(random_day)

df = pd.DataFrame()    
df['birth_date'] =birth_date_list
df['gestational_week'] = gestational_week_list
df['gestational_week_plus_day'] = gestational_week_plus_day_list
df['patient_id'] = df.index
df['patient_id'] = 'patient_'+df['patient_id'].astype(str)
df
```

```python
result_data = {}
for i in range(df.shape[0]):
    date_ = df.birth_date.iloc[i]
    week_ = df.gestational_week.iloc[i]
    patient = df.patient_id.iloc[i]
    result_data.update({patient:premature_reminders(date_,week_)})
reminder_df = pd.DataFrame.from_dict(data=result_data,orient='index')
reminder_df['patient_id'] = reminder_df.index
reminder_df.reset_index(drop=True,inplace=True)
```

```python
df = pd.merge(df,reminder_df,on='patient_id',how='right')
```
